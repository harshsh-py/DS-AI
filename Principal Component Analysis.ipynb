PCA

It is a ststistical procedure to convert a set of observations of possibly corelated variables into a set of values nearly uncorelated variables

Uses:



1. It is used to find the inter-relation between variables in data.
2. Used to interpret and visualize data.
3. Performed on square symmetric matrix.




Objectives:

1. PCA is basically a dimension reduction process but there is no guarantee that the dimensioin is interpretable.
2. Its a non dependent procedure in which it reduces attribute space from large no. of variables to smaller no of factors.
3. The main task is to select a subset of variable from larger set based on whiuch original variables have the highest core;ation with principal amount.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_breast_cancer
```

```python
#importing dataset and creatig dataframes

cancer = load_breast_cancer()
df=pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df)
scaled_data = scaler.transform(df)
```

```python
scaled_data
```

```python
#PCA

from sklearn.decomposition import PCA
PCA= PCA(n_components=2)
PCA.fit(scaled_data)
x_pca = PCA.transform(scaled_data)
x_pca.shape
```

```python
plt.figure(figsize =(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1], c=cancer['target'],cmap='plasma')
plt.xlabel= ('First Principal Component')
plt.ylabel= ('Second Principal Component')
```
