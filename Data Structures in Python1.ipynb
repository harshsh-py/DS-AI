##TOPIC: Lists in Python

Data Structure
It is a way of organizing or storing data




1. List is ordered collection of item
2. It is mutable
3. It allows different type of data element to be present in list.

```python
fruits = ["apple","mango","cherry",1,3.0,2,True,"berry"] #initialise a list
fruits
```

```python
#indexing (to retreive someting from set of elements)
#indexing in python always starts from 0.
fruits[2]
```

```python
#If we want to get the last item in list, we can do : [-1]
fruits[-1]
```

#Append function
It is used to add element in list.

```python
#to add an element into list, we use APPEND

fruits.append("litchy")
```

```python
fruits
```

#Remove function

Used to remove element from list.

```python
fruits.remove(2)
```

```python
fruits
```

```python
fruits.remove("apple")
```

```python
fruits
```

Understanding Iterating in list


```python
for x in fruits:
  print(x)
```

```python
#length
len(fruits)
```

```python
#existence of item
"mango" in fruits
```

```python
"banana" in fruits
```

```python
print("mango" in fruits)
```
