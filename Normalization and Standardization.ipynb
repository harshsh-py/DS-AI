NORMALIZATION (Min-Max)

1. minimum and maximum value of feature are used for scaling
2. Used when feature are on different scale
3. Scales value between [0,1] or [-1,1]
4. Affected by outliers

```python
from sklearn import preprocessing
import numpy as np
import pandas as pd

numpy_array = np.array([2,3,5,7,17,20,80,150])
normalized_array = preprocessing.normalize([numpy_array])
print(normalized_array)
```

STANDARDIZATION (Z- score)

1. Mean and Standard deviation are used for scaling
2. Used when we want to ensure ZERO mean and UNIT standard deviation
3. It is not bounded to range
4. Less affected by outliers.

```python
from sklearn.preprocessing import StandardScaler

data = [[0,0],[0,0],[1,1],[2,2]]
scaler = StandardScaler()
scaler.fit(data)
print(scaler.transform(data))
```
